```python
import pandas as pd
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
ipl= pd.read_csv('matches.csv')
```

```python
ipl.head()
```

```python
ipl.shape
```

```python
ipl['player_of_match'].value_counts() #players with most M.O.M
```

```python
ipl['player_of_match'].value_counts()[0:5] #top 5 M.O.M
```

```python
plt.figure(figsize=(7,7))
plt.bar(list(ipl['player_of_match'].value_counts()[0:5].keys()),list(ipl['player_of_match'].value_counts()[0:5]))
plt.show()
```

```python
ipl['result'].value_counts()  #Results
```

```python
ipl['toss_winner'].value_counts()[0:5]  #Top 5 teams winning most number of toss
```

```python
batting_first=ipl[ipl['win_by_runs']!=0]
```

```python
    batting_first['winner'].value_counts()[0:5]  #Top 5 teams having good record of batting first
```

```python
plt.figure(figsize=(12,7))
plt.bar(list(batting_first['winner'].value_counts()[0:5].keys()),list(batting_first['winner'].value_counts()[0:5]))
plt.show()
```

```python
plt.figure(figsize=(8,13))
plt.pie(list(batting_first['winner'].value_counts()[0:7]),labels=list(batting_first['winner'].value_counts()[0:7].keys()),autopct='%0.1f%%',colors=['blue','yellow','red','purple','red','orange','pink'])
plt.show()
```
